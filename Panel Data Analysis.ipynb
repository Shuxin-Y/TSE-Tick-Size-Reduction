```python
def _panelanalysis(time, para, ee, te, stk_splt1, stk_splt2, bgn1, edn1, bgn2, edn2):
    t = int(time)
    print('                                                 Reading data')
#     pre_1 = pd.read_csv('/Volumes/LaCie/Panel/%smin_Panel_Pre_1.csv'%time)
#     post_1 = pd.read_csv('/Volumes/LaCie/Panel/%smin_Panel_Post_1.csv'%time)
#     phase_1 = pd.concat([pre_1,post_1])
    phase_1 = pd.read_csv('/Volumes/LaCie/Panel/Phase_1_%smin%s%s.csv'%(time,bgn1,edn1))
    mi_p1 = phase_1.set_index(['Code','Time Code'])
    mi_p1['Post 1A'] = mi_p1['Post 1']*mi_p1['Group A']
    
    
#     pre_2 = pd.read_csv('/Volumes/LaCie/Panel/%smin_Panel_Pre_2.csv'%time)
#     post_2 = pd.read_csv('/Volumes/LaCie/Panel/%smin_Panel_Post_2.csv'%time)
#     phase_2 = pd.concat([pre_2,post_2])
    phase_2 = pd.read_csv('/Volumes/LaCie/Panel/Phase_2_%smin%s%s.csv'%(time,bgn2,edn2))
    mi_p2 = phase_2.set_index(['Code','Time Code'])
    mi_p2['Post 2D'] = mi_p2['Post 2']*mi_p2['Group D']
    mi_p2['Post 2E'] = mi_p2['Post 2']*mi_p2['Group E']
    
    print('                                              Starting estimating')
    
    print('\n',
          '                                                                        Fixed Effect Estimation - Phase 1','\n')
#     exog_var = ['Post 1', 'Post 1A', 'Split 1', 'Split 2', 'Lag Return-%s'%time]
    exog_var = ['Post 1', 'Post 1A', 'Lag Return-%s'%time]+stk_splt1
    FE_mod_1 = PanelOLS(mi_p1['%s-%s'%(para, time)], mi_p1[exog_var],entity_effects=ee, time_effects=te)
    print(FE_mod_1.fit())
    
#     print('\n',
#           '                                                                        Random Effect Estimation - Phase 1','\n')
#     exog_var = ['Post 1', 'Post 1A', 'Split 1', 'Split 2', 'Lag Return-%s'%time]
#     RE_mod_1 = RandomEffects(mi_p1['%s-%s'%(para, time)], mi_p1[exog_var])
#     print(RE_mod_1.fit())
    
    print('\n',
          '                                                                        Fixed Effect Estimation - Phase 2','\n')
#     exog_var = ['Post 2','Post 2D','Post 2E', 'Split 2', 'Split 3', 'Lag Return-%s'%time]
    exog_var = ['Post 2','Post 2D','Post 2E', 'Lag Return-%s'%time]+stk_splt2
    FE_mod_2 = PanelOLS(mi_p2['%s-%s'%(para, time)], mi_p2[exog_var],entity_effects=ee, time_effects=te)
    print(FE_mod_2.fit())

#     print('\n',
#           '                                                                        Random Effect Estimation - Phase 2','\n')
#     exog_var = ['Post 2','Post 2D','Post 2E', 'Split 2', 'Split 3', 'Lag Return-%s'%time]
#     RE_mod_2 = RandomEffects(mi_p2['%s-%s'%(para, time)], mi_p2[exog_var])
#     print(RE_mod_2.fit())
```
