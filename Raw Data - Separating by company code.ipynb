```python
# Import Modules
import pandas as pd
import numpy as np
import math

import os
import warnings
warnings.filterwarnings('ignore')

import jupyternotify
ip = get_ipython()
ip.register_magics(jupyternotify.JupyterNotifyMagics(ip))
```

```python
co_list = pd.read_csv('/Volumes/LaCie/TOPIX100_list_2014.csv')['Code']
co_list = co_list.sort_values().reset_index(drop=True)
```

```python
def _rawprocess(year, month):
    data = pd.read_csv('/Volumes/LaCie/Data/stock_tick_%s_%s.csv'%(year,month))
    dates = data['Date'].unique()
    for j in dates:
        tmp_1 = data[data['Date']==j]
        for k in co_list:
            print ('%s-%s-%s-%s'%(str(j)[:4], str(j)[4:6], str(j)[-2:], str(k)))
            tmp = tmp_1[tmp_1['Issue code']==k]
            path = '/Volumes/LaCie/aTransactions_%s/%s/%s'%(str(j)[:4], str(j)[4:6], str(j)[-2:])
            if not os.path.exists(path):
                os.makedirs(path)
            pd.DataFrame(tmp).to_csv('/Volumes/LaCie/aTransactions_%s/%s/%s/%s.csv'%
                                    (str(j)[:4], str(j)[4:6], str(j)[-2:], str(k)),
                                     index=False)
```

```python
%%notify
for i in ['01','02','03','04','05','06','07','08','09','10','11','12']:
    _rawprocess('2014',i)
```

```python
%%notify
for i in ['07','08','09','10','11','12']:
    _rawprocess('2013',i)
```
